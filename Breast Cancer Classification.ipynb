# MODEL TRAINING

```python
# Deleting the target value because on training we don't want
x = df_cancer.drop(['target'], axis = 1)
```

```python
x
```

```python
y = df_cancer['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# For random split of x and y data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model_train = SVC()
```

```python
svc_model_train.fit(X_train, y_train)
```

# EVALUATING THE MODEL

```python
y_predict = svc_model_train.predict(X_test)
```

```python
# Model not trained (Like Useless)
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot= True)
```

# IMPROVING THE MODEL

```python
min_train = X_train.min()
```

```python
range_train = (X_train - min_train).max()
```

```python
X_train_scaled = (X_train - min_train) / range_train
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test) / range_test
```

```python
svc_model_train.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model_train.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# IMPROVING THE MODEL MORE

```python
param_grid = {'C': [0,1,10,100], 'gamma': [1,0.1,0.01, 0.001], 'kernal': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled, y_train)
```
